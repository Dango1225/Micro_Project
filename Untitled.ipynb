```python
DATABASE = 'micro_p3'
import mysql.connector
import csv

# File read function of .csv file. Use yield method to return a iterator, such that the memory spend will always be 1 row each time. 
def read_file(fpath):
    with open(fpath, 'r') as csvfile:
        data = csv.reader(csvfile)
        for row in data:
            yield(row)

def create_table(fname, db, cur, small = False):
    
    fpath = fname + '.csv'

    # Save the basic information (header, nrows and ncols) and pick a example row. 
    # Inside the iteration, we need to figure out the maximum element length of each row, which updated as eleLen.
    # We need to delete all the spaces at the beginning and the end of each element.
    nrows = 0
    for i, row in enumerate(read_file(fpath)):
        thisRow = list(row)
        thisRow = [x.strip() for x in thisRow]
        if i == 0:
            header = thisRow
            header = ['_'.join(x.split(' '))[:60] for x in header]  #Remove the spaces in header
            ncols = len(header)
        if i == 1:
            exampleRow = thisRow
            eleLen = [len(exampleRow[i]) for i in range(ncols)]
        if i > 1:
            for i in range(ncols):
                if len(exampleRow[i]) < len(thisRow[i]):
                    eleLen[i] = len(thisRow[i])
                if not exampleRow[i]:
                    exampleRow[i] = list(row)[i]
        nrows += 1

    # Create the datatype of exampleRow. 0 for varchar, 1 for int, 2 for float, -1 for NULL, then 0 for other type as char.
    dataType = [];
    for i, ele in enumerate(exampleRow):
        if not ele:    # Auto delete all-NULL column
            header.pop(i)
            ncols -= 1
        dataType.append(0)   
        if ele[-1] == '%':
            dataType[-1] = 2
        elif ele[0].isdigit() or ele[0] == '$':
            if '.' in ele:
                dataType[-1] = 2
            else:
                dataType[-1] = 1

    print('Example: ', exampleRow)
    print('Maximum element length: ', eleLen)

    # Datatype to SQL commands
    cols_cmd = []
    cols_cmd.append('`auto_id` int(11) NOT NULL AUTO_INCREMENT,')
    for i, typ in enumerate(dataType):
        result = '`%s` ' %(header[i])
        if typ == 0:
            eleLen[i] = eleLen[i] * 2 + 1
            tempStr = 'varchar(%d),' %(eleLen[i])
        if typ == 1:
            eleLen[i] = eleLen[i] + 1
            tempStr = 'int(%d),' %(eleLen[i])
        if typ == 2:
            tempStr = 'double(%d,%d),' %(eleLen[i], 2)
        result = result + tempStr
        cols_cmd.append(result)
    primary_cmd = 'PRIMARY KEY (`auto_id`)'

    table_create_cmd = 'CREATE TABLE `%s` (%s) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4 COLLATE=utf8mb4_0900_ai_ci;' %(fname, ''.join(cols_cmd) + primary_cmd)  # The command line to create the table

    # Create the form 'patient_churn'
    
    cur.execute('DROP DATABASE IF EXISTS `%s`' %(DATABASE))
    cur.execute('CREATE DATABASE `%s`' %(DATABASE))
    cur.execute('USE `%s`' %(DATABASE))
    cur.execute(table_create_cmd)
    for i, row in enumerate(read_file(fpath)):
        if small and i == 10000: break
        if i == 0: continue  # Skip the head row
        thisRow = list(row)
        if not any(thisRow): continue   # Skip the all-NULL row
        for j in range(ncols):
            thisRow[j] = thisRow[j].strip()
            thisRow[j] = thisRow[j].lstrip('$')
            thisRow[j] = thisRow[j].rstrip('%')
            if not thisRow[j]:
                thisRow[j] = 'NULL'
            elif dataType[j] >0:
                thisRow[j] = ''.join(thisRow[j].split(','))
            elif dataType[j] == 0:
                thisRow[j] = '\'' + thisRow[j] + '\''
        insert_cmd = 'INSERT INTO `%s` VALUES (%s,%s);' %(fname, str(i), ','.join(thisRow))
        cur.execute(insert_cmd)
    
    return [nrows, ncols, header, exampleRow, eleLen]
```

```python
# GOAL 2: The statistics.
import matplotlib.pyplot as plt
import numpy as np
  
def fetch_datatype(cur):
    types = []
    for item in cur.description:
        if item[1] == 3:
            types.append('int')
        elif item[1] == 5:
            types.append('double')
        elif item[1] == 253:
            types.append('char')
        else:
            types.append('unknown')
    return types

def fetch_EDA(cur, fname, col, names = ['max', 'min', 'mean', 'median', 'std', 'nNULLs', 'card']):
    res = []
    for name in names:
        if name == 'max':
            cur.execute('SELECT MAX(%s) FROM %s WHERE %s IS NOT NULL' %(col, fname, col))
            res.append(cur.fetchall()[0][0])
        if name == 'min':
            cur.execute('SELECT MIN(%s) FROM %s WHERE %s IS NOT NULL' %(col, fname, col))
            res.append(cur.fetchall()[0][0])
        if name == 'mean':
            cur.execute('SELECT AVG(%s) FROM %s WHERE %s IS NOT NULL' %(col, fname, col))
            res.append(float(cur.fetchall()[0][0]))
        if name == 'median':
            cur.execute('SELECT x.%s from %s x, %s y \
                    GROUP BY x.%s \
                    HAVING SUM(SIGN(1-SIGN(y.%s-x.%s)))/COUNT(*) > .5 \
                    LIMIT 1' %(col, fname, fname, col, col, col))
            res.append(float(cur.fetchall()[0][0]))
        if name == 'std':
            cur.execute('SELECT STD(%s) FROM %s WHERE %s IS NOT NULL' %(col, fname, col))
            res.append(float(cur.fetchall()[0][0]))
        if name == 'nNULLs':
            cur.execute('SELECT COUNT(IFNULL(%s, 1)) FROM %s WHERE %s IS NULL' %(col, fname, col))
            res.append(cur.fetchall()[0][0])
        if name == 'card':
            cur.execute('SELECT COUNT(DISTINCT(%s)) FROM %s' %(col, fname))
            res.append(cur.fetchall()[0][0])
    return res

def fetch_corr(cur, fname, col1, col2):
    cur.execute('SELECT (Count(*)*Sum(d.%s*d.%s)-Sum(d.%s)*Sum(d.%s)) / \
         (sqrt(Count(*)*Sum(d.%s*d.%s)-Sum(d.%s)*Sum(d.%s)) * \
          sqrt(Count(*)*Sum(d.%s*d.%s)-Sum(d.%s)*Sum(d.%s))) AS TotalCorelation \
          FROM %s d' %(col1, col2, col1, col2, col1, col1, col1, col1, col2, col2, col2, col2, fname))
    return cur.fetchall()[0][0]

db = mysql.connector.connect( host="localhost", user="root", passwd="951225")
cur = db.cursor()

# Create the table first
fname = 'Covid_expenditures_2020'
[nrows, ncols, header, exampleRow, eleLen] = create_table(fname, db, cur)
cur.execute('USE `%s`' %(DATABASE))
cur.execute('SELECT * FROM %s' %(fname))

# Fetch the datatype
print('datatype: ', fetch_datatype(cur))
cur.fetchall()

# Fetch the EDAs
print('header:', header)
res = fetch_EDA(cur, fname, header[6])
print('Statistics: ', res)

# Fetch the correlations
corr = [[0] * 4 for _ in range(4)]
for i in range(4):
    for j in range(4):
        corr[i][j] = fetch_corr(cur, fname, header[i + 3], header[j + 3])

cur.close()
db.close()
```

```python
# Plot the correlations
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Construct arrays for the anchor positions of the 16 bars.
xpos, ypos = np.meshgrid([1,2,3,4], [1,2,3,4], indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

# Construct arrays with the dimensions for the 16 bars.
dx = dy = 0.3
dz = sum(corr, [])

cmap = cm.get_cmap('jet')
rgba = [cmap((k-min(dz))/max(dz)) for k in dz] 
ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color=rgba, zsort='average')

ax.set_xticks([1,2,3,4])
ax.set_xticklabels(['reg','cvd','i_cvd','ip_rate'])
ax.set_yticks([2,3,4,5])
ax.set_yticklabels(['reg','cvd','i_cvd','ip_rate'])

ax.set_xlabel('factor 1')
ax.set_ylabel('factor 2')
ax.set_zlabel('correlation')
ax.set_title('Correlation Map')

plt.show()
```

```python
# Example: patient_churn data
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd

db = mysql.connector.connect( host="localhost", user="root", passwd="951225")
cur = db.cursor()

# Create the table first
fname = 'patient_churn'
[nrows, ncols, header, exampleRow, eleLen] = create_table(fname, db, cur, small = True)
print(header)

# Get some statictics from the table
stats = []
for head in header:
    if head != 'Death_Year':
        stats.append(fetch_EDA(cur, fname, head, ['max','min','mean','card']))
stats_reason = stats[-6:]
print(stats_reason)

fig = plt.figure()
ax = fig.add_subplot()
ax.plot([stats_reason[i][2] for i in range(len(stats_reason))])
ax.set_xlabel('Reason')
ax.set_ylabel('Non-zero percentage')
plt.show()

cur.execute('SELECT * FROM %s' %(fname))
allData = []
allLabel = []
seen = {}
while True:
    row = cur.fetchone()
    if not row: break
    i, *item = row
    if any(item[-6:]):
        temp = []
        temp.append(item[1])  # Sex: 1 for male, 2 for female
        temp.append(2015 - item[2])   # Calcuate age from birth year
        temp.append(True if item[3] else False)  # Death: True or False
        temp.append(item[4])
        temp.append(item[5])
        allData.append(temp)
        allLabel.append(''.join(['y' if x else 'n' for x in item[-6:]]))

allDataPd = pd.DataFrame(allData)
allLabelPd = pd.DataFrame(allLabel)
xTrain, xTest, yTrain, yTest = train_test_split(allDataPd, allLabelPd, test_size = 0.2)
model = DecisionTreeClassifier(max_depth=3)
model.fit(xTrain, yTrain)
predictions = model.predict(xTest)
score = accuracy_score(yTest, predictions)
print(score)

cur.close()
db.close()
```
